```python
def parse_string(input_string: str) -> dict:
    result = {}
    if input_string.strip() == "good": 
        result["good"] = "good"
        return result
    
    lines = input_string.strip().split('  \n')

    for line in lines:
        key_value = line.split(': ', 1)
        if len(key_value) == 2:
            key, value = key_value
            result[key.strip()] = value.strip()

    return result
```
